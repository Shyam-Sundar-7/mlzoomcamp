# Analysis with target

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a cross-tabulation table
cross_tab = pd.crosstab(df['Vehicle_Age'], df['Target'], normalize='index') * 100

# Plot the cross-tabulation table as a bar plot
cross_tab.plot(kind='bar')
plt.xlabel('Vehicle Age')
plt.ylabel('Percentage')
plt.title('Percentage of Target for each Vehicle Age')

plt.show()
```

After 1 year, we have higher probability that the customer is interested in opting fot the Vehicle Insurance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the bar plot
sns.countplot(x='Previously_Insured', hue='Target',data=df)

# Set labels and title
plt.xlabel('Previously Insured')
plt.ylabel('Count')
plt.title('Bar Plot of Previously Insured and Target')

# Show the plot

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the box plot
sns.boxplot(x='Target', y='Annual_Premium', data=df)

# Set labels and title
plt.xlabel('Target')
plt.ylabel('Annual Premium')
plt.title('Box Plot of Annual Premium and Target')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the percentage of each 'Target' category within each 'Gender' category
percentage_df = df.groupby(['Gender', 'Target']).size().unstack() / df.groupby('Gender').size().values[:, None] * 100

# Create the bar plot
percentage_df.plot(kind='bar')

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.title('Percentage of Target by Gender')

# Show the plot
plt.show()
```

    Males prefer to go for insurance. But still there isn't big difference between genders.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the cross-tabulation table and normalize the values
cross_tab = pd.crosstab(df['Driving_License'], df['Target'], normalize='index') * 100

# Plot the cross-tabulation table as a bar plot
cross_tab.plot(kind='bar')

# Set labels and title
plt.xlabel('Driving License')
plt.ylabel('Percentage')
plt.title('Percentage of Target by Driving License')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the count plot
sns.countplot(y='Region_Code', hue='Target', data=df)

# Set labels and title
plt.ylabel('Region Code')
plt.xlabel('Count')
plt.title('Count Plot of Region Code and Target')

# Show the plot
plt.show()
```

Equal importance between the Region_Code and Target

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the count plot
sns.countplot(x='Vehicle_Damage', hue='Target', data=df)

# Set labels and title
plt.xlabel('Vehicle Damage')
plt.ylabel('Count')
plt.title('Count Plot of Vehicle Damage and Target')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the count plot
sns.countplot(y='Policy_Sales_Channel', hue='Target', data=df)

# Set labels and title
plt.ylabel('Policy_Sales_Channel')
plt.xlabel('Count')
plt.title('Count Plot of Policy_Sales_Channel and Target')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
df['MALE']=df['Gender'].apply(lambda x: 1 if x=='Male' else 0)
df['vehicle_d']=df['Vehicle_Damage'].apply(lambda x: 1 if x=='Yes' else 0)
corr_matrix = df[['Age','MALE', 'Annual_Premium', 'Vintage','Driving_License', 'Previously_Insured', 'Policy_Sales_Channel','Target','vehicle_d']].corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Set title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```
